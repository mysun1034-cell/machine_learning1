```python
import numpy as np
```

```python
scores = np.array([80, 90, 70, 85])
scores + 10       # 전체에 더하기
scores * 1.1      # 10% 가산
scores.mean()     # 평균
scores.max()      # 최댓값
```

```python
arr = np.array([80, 90, 70])
arr.dtype   # int64
arr.shape   # (3,)
arr.ndim    # 1

arr2 = np.array([[1, 2, 3],
                 [4, 5, 6]])
arr2.shape  # (2, 3) — 2행 3열

# 인덱싱
arr2[0, 1]  # 0행 1열 → 2
arr2[:, 0]  # 모든 행의 0열
arr2[1, :]  # 1행 전체
```
