```python
import pandas as pd
TITANIC_URL = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
```

```python
# 데이터 불러오기
titanic = pd.read_csv(TITANIC_URL)
```

```python
#데이터의 정보를 확인한다.
titanic.info()
```

```python
titanic.head
```

```python
titanic.loc[0]
```

```python
titanic.iloc[0:3]
```

```python
#Q. pandas에서 iloc와 loc의 차이에 대해서 알아보기.
```

```python
#조건부로 필터링하기
titanic[(titanic['Survived'] == 1) & (titanic['Sex'] == 'female')]
#컬럼명 정확히. 대소문자 구분, 각각의 조건식마다 () 구분, and가 아니라 & 사용
```

```python
#결측치
titanic.isnull().sum() # 컬럼별 결측치(조건이 True)의 합을 확인
```

```python
#결측치 채우기
titanic.copy()
```

```python
df = titanic.copy()#사본 만들기
#df['age']의 na(빈값)을 'age'의 중간값으로 채워라.
df['Age'] = df['Age'].fillna(df['Age'].median())
#비어있는 값을 다 날리기
df['Age'] = df['Age'].dropna()
```

```python
df.isnull().sum()
```

```python
### groupby 그룹별로 보기
df.groupby('Sex')["Survived"].mean()
# 그룹화 한 함수의 결과(object)에 집계를 함
```

```python
#집계 함수
df.groupby('Pclass').agg(생존율=('Survived', 'mean'), 인원=('Survived', 'count'))
```

```python
#남녀성비를 비율로
df['Sex'].value_counts(normalize=True)
```

```python
# 숫자로 카운팅
df['Sex'].value_counts
```
