```python
df = pd.read_csv('titanic.csv')

X = df[['Pclass']].values   # 문제지 (Pclass 열만)
y = df['Survived'].values   # 정답지 (생존 여부)
```

```python
train_x, test_x, train_y, test_y = train_test_split(
    X, y,
    random_state=42,
    stratify=y
)
```

```python
kn = KNeighborsClassifier()
kn.fit(train_x, train_y) 
```

```python
print(kn.predict([[1]]))   # 1등석
print(kn.predict([[2]]))   # 2등석
print(kn.predict([[3]]))   # 3등석
```

```python
train_score = kn.score(train_x, train_y)
test_score  = kn.score(test_x, test_y)
```

```python
print(f"훈련 정확도: {train_score:.4f}")
print(f"테스트 정확도: {test_score:.4f}")
```

```python
df = pd.read_csv('titanic.csv')

# 등급별 데이터 분리
class1 = df[df['Pclass'] == 1]
class2 = df[df['Pclass'] == 2]
class3 = df[df['Pclass'] == 3]

plt.figure(figsize=(8, 6))
plt.scatter(class1['Pclass'], class1['Survived'], marker='^', label='1등석')
plt.scatter(class2['Pclass'], class2['Survived'], marker='o', label='2등석')
plt.scatter(class3['Pclass'], class3['Survived'], marker='s', label='3등석')
plt.xlabel('객실 등급')
plt.ylabel('생존 여부 (1=생존, 0=사망)')
plt.title('객실 등급별 생존 여부')
plt.legend()
plt.show()
```
